```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([[i]*10 for i in xrange(1,51,5)]).flatten()
y = np.array([i for i in xrange(1,51,5)]*10)
x = [i+np.random.uniform(-1,1) for i in x]
y = [i+np.random.uniform(-1,1) for i in y]
nodes = zip(x,y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def distance(x,y):
    return ((x[0]-y[0])**2+(x[1]-y[1])**2)**0.5

def return_neighbours(point,graph):
    min_dist=6
    n = []
    for node in graph:
        if node!=point and distance(point,node)<=min_dist:
            n.append(node)
    return n
```

```python
edges = list()
edge_set = list()
for node in nodes:
    neighbours = return_neighbours(node,nodes)
    for neighbour in neighbours:
        node_prob = np.random.uniform(0,1)
        neighbour_count = edge_set.count(neighbour)*0.5
        neighbour_prob = np.random.uniform(0,1)
        #print neighbour_count
        prob = (neighbour_count*neighbour_prob)+node_prob
        print prob
        if prob>0.7:
            edges.append([node,neighbour])
            edge_set.append(neighbour)
```

```python
import matplotlib
import matplotlib.collections

lc = matplotlib.collections.LineCollection(edges)

fig, ax = plt.subplots()
ax.add_collection(lc)
plt.plot(x,y,'o')

plt.show()
```
